# Decision Tree Classifier

```python
#import data processing library
import pandas as pd
import numpy as np
```

```python
#import visualization library
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import the data
customers=pd.read_csv('http://datasets.coso.io/public/Car_Customers.csv')
```

```python
customers.head()
```

```python
#create distribution plot
sns.distplot(customers['EstimatedSalary'])
```

```python
sns.pairplot(customers,hue='Gender')
```

```python
#training set and test data set
from sklearn.model_selection import train_test_split
```

```python
X=customers.iloc[:,[2,3]].values
y=customers.iloc[:,[4]].values
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.30, random_state=50)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy', random_state=50
                         )
dt.fit(X_train,y_train)
```

```python
#predictions
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
```

```python
print(cm)
print(cr)
```

```python
from matplotlib.colors import ListedColormap
```

```python
X_set,y_set=X_test,y_test
X1,X2=np.meshgrid(np.arrange(start=X_set[:]))
```

```python

for i , j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j],1),
    c=ListedColormap(('red','green'))(i),label=j)
plt.title('Decision Tree Classification')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
